# Collapsed coordinate transformation in 3D

```julia
using StableSpectralElements, Plots, LaTeXStrings, Plots.PlotMeasures
p = 6
ref_tet = ReferenceApproximation(
    NodalTensor(p), Tet(), volume_quadrature_rule=Tuple(LGQuadrature(p) for m in 1:3),
    facet_quadrature_rule=Tuple(LGLQuadrature(p) for m in 1:2));
ref_hex = ReferenceApproximation(
    NodalTensor(p), Hex(), volume_quadrature_rule=LGQuadrature(p), 
    facet_quadrature_rule=LGQuadrature(p));
```

```julia
p_tet = plot(windowsize=(350,300), ref_tet, sketch=false, volume_quadrature=false, facet_quadrature=true,
    facet_connect=true, facet_inds=[3,4,1,2], xlims=[-1.1,1.1],ylims=[-1.1,1.1],zlims=[-1.1,1.1],
    xticks=([-1.0, 0.0, 1.0], [L"$-1.0$",L"0.0",L"1.0"]), 
    yticks=([-1.0, 0.0, 1.0], [L"$-1.0$",L"0.0",L"1.0"]),
    zticks=([-1.0, 0.0, 1.0], [L"$-1.0$",L"0.0",L"1.0"]),
    xlabelfontsize=16, ylabelfontsize=16, zlabelfontsize=16,camera=(60,22.5),
    xlabel=L"$\xi{}_1$",ylabel=L"$\xi{}_2$", zlabel=L"$\xi{}_3$",
    xtickfontsize=12, xtickfontvalign=:top,
    ytickfontsize=12,  ytickfontvalign=:top,
    ztickfontsize=12,
    facet_color_inds=[1,2,:orangered4,:slateblue4],
    grid_line_width=3, edges=false, redraw_edge=false)
```

```julia
p_hex = plot()
p_hex = plot(windowsize=(350,300),ref_hex, sketch=false, volume_quadrature=false, facet_quadrature=true,
    facet_connect=true,  xlims=[-1.1,1.1],ylims=[-1.1,1.1],zlims=[-1.1,1.1],
    facet_inds=[1,5,3,4,6,2], facet_color_inds=[1,2,:orangered4,3,:darkorchid4,:slateblue4],
    xticks=([-1.0, 0.0, 1.0], [L"$-1.0$",L"0.0",L"1.0"]), 
    yticks=([-1.0, 0.0, 1.0], [L"$-1.0$",L"0.0",L"1.0"]),
    zticks=([-1.0, 0.0, 1.0], [L"$-1.0$",L"0.0",L"1.0"]),
    xlabelfontsize=16, ylabelfontsize=16, zlabelfontsize=16,camera=(60,22.5),
    xlabel=L"$\eta{}_1$",ylabel=L"$\eta{}_2$", zlabel=L"$\eta{}_3$",
    xtickfontsize=12, xtickfontvalign=:top,
    ytickfontsize=12,  ytickfontvalign=:top,
    ztickfontsize=12,
    grid_line_width=3,edges=false, redraw_edge=false)
```
