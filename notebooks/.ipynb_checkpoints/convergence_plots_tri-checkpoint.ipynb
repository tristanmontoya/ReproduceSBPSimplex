```julia
using CLOUD, JLD2, Plots, RecipesBase, LaTeXStrings
```

```julia
path = "results/20230422/"
```

```julia
analysis = []
results = []

initial_data = InitialDataSine(1.0,Tuple(2*π for m in 1:2))

ref_path = string(path, "NodalMulti_Tri_SkewSymmetricMapping_p4/central/")
push!(analysis, RefinementErrorAnalysis(initial_data,ref_path))
push!(results, analyze(last(analysis)))
ref_path = string(path, "NodalMulti_Tri_SkewSymmetricMapping_p4/upwind/")
push!(analysis, RefinementErrorAnalysis(initial_data,ref_path))
push!(results, analyze(last(analysis)))
ref_path = string(path, "ModalMulti_Tri_SkewSymmetricMapping_p4/central/")
push!(analysis, RefinementErrorAnalysis(initial_data,ref_path))
push!(results, analyze(last(analysis)))
ref_path = string(path, "ModalMulti_Tri_SkewSymmetricMapping_p4/upwind/")
push!(analysis, RefinementErrorAnalysis(initial_data,ref_path))
push!(results, analyze(last(analysis)))
```

```julia
ref_path = string(path, "NodalTensor_Tri_SkewSymmetricMapping_p4/central/")
push!(analysis, RefinementErrorAnalysis(initial_data,ref_path))
push!(results, analyze(last(analysis)))
ref_path = string(path, "NodalTensor_Tri_SkewSymmetricMapping_p4/upwind/")
push!(analysis, RefinementErrorAnalysis(initial_data,ref_path))
push!(results, analyze(last(analysis)))
ref_path = string(path, "ModalTensor_Tri_SkewSymmetricMapping_p4/central/")
push!(analysis, RefinementErrorAnalysis(initial_data,ref_path))
push!(results, analyze(last(analysis)))
ref_path = string(path, "ModalTensor_Tri_SkewSymmetricMapping_p4/upwind/")
push!(analysis, RefinementErrorAnalysis(initial_data,ref_path))
push!(results, analyze(last(analysis)))
```

```julia
h = [1.0/2^n for n in eachindex(results[1].dof[:,2])]
plt_m = plot(xaxis=:log10, yaxis=:log10, windowsize=(400,300), legendfontsize=9,
    fontfamily="Computer Modern", xtickfontsize=10, ytickfontsize=10, 
    xguidefontsize=12, yguidefontsize=12,
    ylabel="\$L^2\$ error", xlabel="Element size \$h\$", legend=:topleft, xlims=[2e-2,1.0],  ylims=[5e-8,10e-1])
plot!(plt_m, h, results[1].error, label="", 
    linewidth=3.0, linestyle=:dash, markershape=:circle, markersize=7, markerstrokewidth=0.0, color=:black)
plot!(plt_m, h,  results[2].error, label="Nodal multidimensional", 
    linewidth=3.0, linestyle=:solid, markershape=:circle, markersize=7, markerstrokewidth=0.0, color=:black)
plot!(plt_m, h,  results[3].error, label="", 
    linewidth=2.0, linestyle=:dash, markershape=:circle, markersize=4, markerstrokewidth=0.0, color=palette(:default)[1])
plot!(plt_m, h ,  results[4].error, label="Modal multidimensional", 
    linewidth=2.0, linestyle=:solid, markershape=:circle, markersize=4, markerstrokewidth=0.0, color=palette(:default)[1])
r1 = h[end-1]
r2 = h[end-2]
plot!(plt_m, [r1, r2, r2, r1], 3.5*[r1^5, r2^5, r1^5, r1^5], 
    label=:none, linewidth=2, color=:black,  annotations = (0.082, 6.5*r1^5, L"\mathbf{5:1}", 12))
```

```julia
h = [1.0/2^n for n in eachindex(results[1].dof[:,2])]
plt_t = plot(xaxis=:log10, yaxis=:log10, windowsize=(400,300), legendfontsize=9,
    fontfamily="Computer Modern", xtickfontsize=10, ytickfontsize=10, 
    xguidefontsize=12, yguidefontsize=12,
    ylabel="\$L^2\$ error", xlabel="Element size \$h\$", legend=:topleft, xlims=[2e-2,1.0],  ylims=[5e-8,10e-1])
plot!(plt_t, h, results[5].error, label="", 
    linewidth=3.0, linestyle=:dash, markershape=:square, markersize=7, markerstrokewidth=0.0, color=:black)
plot!(plt_t, h,  results[6].error, label="Nodal tensor-product", 
    linewidth=3.0, linestyle=:solid, markershape=:square, markersize=7, markerstrokewidth=0.0, color=:black)
plot!(plt_t, h,  results[7].error, label="", 
    linewidth=2.0, linestyle=:dash, markershape=:square, markersize=4, markerstrokewidth=0.0, color=palette(:default)[1])
plot!(plt_t, h ,  results[8].error, label="Modal tensor-product", 
    linewidth=2.0, linestyle=:solid, markershape=:square, markersize=4, markerstrokewidth=0.0, color=palette(:default)[1])
r1 = h[end-1]
r2 = h[end-2]
plot!(plt_t, [r1, r2, r2, r1], 3.5*[r1^5, r2^5, r1^5, r1^5], 
    label=:none, linewidth=2, color=:black,  annotations = (0.082, 6.5*r1^5, L"\mathbf{5:1}", 12))
```

```julia
savefig(plt_m, "tri_p4_multi_conv.pdf")
savefig(plt_t, "tri_p4_tensor_conv.pdf")
```
