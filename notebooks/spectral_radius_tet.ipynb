```julia
using CLOUD, JLD2, Plots, LaTeXStrings
```

```julia
path = "../results/spectral_radius/"

s_nmc = load_object(string(path,"NodalMulti_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_nmu = load_object(string(path,"NodalMulti_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_mmc = load_object(string(path,"ModalMulti_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_mmu = load_object(string(path,"ModalMulti_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_ntc = load_object(string(path,"NodalTensor_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_ntu = load_object(string(path,"NodalTensor_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_mtc = load_object(string(path,"ModalTensor_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_mtu = load_object(string(path,"ModalTensor_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
```

```julia
plt_m_16 = plot(legend=:topleft, xlabel="Polynomial degree \$p\$", xscale=:log10, yscale=:log10,
    ylabel="Spectral radius", fontfamily="Computer Modern", legendfontsize=9,xlims=[1.9,17], ylims=[5*10^1,10^3],
    xguidefontsize=10, yguidefontsize=10, xtickfontsize=10, ytickfontsize=10,
xticks=([2,4,8,16], string.([2,4,8,16])), size=(400,300))
plot!(plt_m_16, collect(2:10), s_nmu[1:9], label="Nodal multidimensional",  markershape=:circle, markersize=5, linewidth=3.0, markerstrokewidth=0.0, color=:black)
plot!(plt_m_16, collect(2:10), s_nmc[1:9], label="", linestyle=:dash, markershape=:circle, markersize=5, linewidth=3.0, markerstrokewidth=0.0, color=:black)
plot!(plt_m_16, collect(2:10), s_mmu[1:9], label="Modal multidimensional",  markershape=:circle, markersize=3, linewidth=2.0, markerstrokewidth=0.0, color=palette(:default)[1])
plot!(plt_m_16, collect(2:10), s_mmc[1:9], label="", linestyle=:dash, markershape=:circle, markersize=3, linewidth=2.0, markerstrokewidth=0.0, color=palette(:default)[1])
r1 = 8
r2 = 10
plot!(plt_m_16, [r1, r2, r2, r1], 4.7*[r1^2, r2^2, r1^2, r1^2], label=:none, linewidth=2, color=:black, annotations = ([9.15], [2.65*10^2], L"\mathbf{2:1}", 10))
```

```julia
plt_t_16 = plot(legend=:topleft, xlabel="Polynomial degree \$p\$", xscale=:log10, yscale=:log10,
    ylabel="Spectral radius", fontfamily="Computer Modern", legendfontsize=9, ylims=[10^1, 6*10^5],
    xguidefontsize=10, yguidefontsize=10, xtickfontsize=10, ytickfontsize=10,
xticks=([2,4,8,16,25], string.([2,4,8,16,25])), size=(400,300))
plot!(plt_t_16, collect(2:8), s_ntu[1:7], label="Nodal tensor-product",  markershape=:square, markersize=3, linewidth=3.0, markerstrokewidth=0.0, color=:black)
plot!(plt_t_16, collect(2:8), s_ntc[1:7], label="", linestyle=:dash, markershape=:square, markersize=3, linewidth=3.0, markerstrokewidth=0.0, color=:black)
plot!(plt_t_16, collect(2:12), s_mtu[1:11], label="Modal tensor-product",  markershape=:square, markersize=3, linewidth=3.0, markerstrokewidth=0.0, markerstrokecolor=palette(:default)[2], color=palette(:default)[2])
plot!(plt_t_16, collect(2:12), s_mtc[1:11], label="", linestyle=:dash, markershape=:square, markersize=3, linewidth=3.0, markerstrokewidth=0.0, markerstrokecolor=palette(:default)[2], color=palette(:default)[2])
r1 = 12
r2 = 16
plot!(plt_m_16, [r1, r2, r2, r1], 4.7*[r1^2, r2^2, r1^2, r1^2], label=:none, linewidth=2, color=:black, annotations = ([9.15], [2.65*10^2], L"\mathbf{2:1}", 10))
plot!(plt_t_16, [r1, r2, r1, r1], 3*[r1^4, r2^4, r2^4, r1^4], label=:none, linewidth=2, color=:black, annotations = ([14.25], [3*10^5], L"\mathbf{4:1}", 10))
```

```julia
savefig(plt_m_16, "../plots/spectral_radius/spectral_radius_tri_multi_2_16.pdf")
savefig(plt_t_16, "../plots/spectral_radius/spectral_radius_tri_tensor_2_16.pdf")
```
