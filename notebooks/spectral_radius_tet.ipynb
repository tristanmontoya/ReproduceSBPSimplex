# Spectral radius for tetrahedra

```julia
using StableSpectralElements, JLD2, Plots, LaTeXStrings
```

```julia
path = "../results/spectral_radius/"

s_nmc = load_object(string(path,"NodalMulti_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_nmu = load_object(string(path,"NodalMulti_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_mmc = load_object(string(path,"ModalMulti_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_mmu = load_object(string(path,"ModalMulti_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_ntc = load_object(string(path,"NodalTensor_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_ntu = load_object(string(path,"NodalTensor_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_mtc = load_object(string(path,"ModalTensor_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_mtu = load_object(string(path,"ModalTensor_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"));
```

```julia
plt_nodal = plot(legend=:topleft, xlabel="\nPolynomial degree \$p\$", xscale=:log10, yscale=:log10,
    ylabel="Spectral radius\n", fontfamily="Computer Modern", legendfontsize=10, ylims=[10^1, 2*10^7],
    xguidefontsize=12, yguidefontsize=12, xtickfontsize=12, ytickfontsize=12,
    xticks=([2,4,8,16,25], string.([2,4,8,16,25])),
    yticks=([10^2,10^4, 10^6], [L"10^2", L"10^4", L"10^6" ]), size=(450,300))

plot!(plt_nodal, collect(2:16), s_ntu[1:15], label="Nodal tensor-product",
    markershape=:square, markersize=3.5, linewidth=2.0, 
    markerstrokewidth=0.0, color=palette(:default)[2])
plot!(plt_nodal, collect(2:16), s_ntc[1:15], label="", 
    linestyle=:dot, markershape=:square, markersize=3.5, linewidth=2, 
    markerstrokewidth=0.0, color=palette(:default)[2])
plot!(plt_nodal, collect(2:10), s_nmu[1:9], label="Nodal multidimensional", 
    markershape=:circle, markersize=3, linewidth=2, markerstrokewidth=0.0,
    markerstrokecolor=:slateblue4, color=:slateblue4)
plot!(plt_nodal, collect(2:10), s_nmc[1:9], label="", 
    linestyle=:dot, markershape=:circle, markersize=3, 
    linewidth=2, markerstrokewidth=0.0, markerstrokecolor=:slateblue4, color=:slateblue4)

r1 = 8
r2 = 10
plot!(plt_nodal, [r1, r2, r1, r1], 9*[r1^2, r2^2, r2^2, r1^2], 
        label=:none, linewidth=2, color=:black, annotations = ([9], [1.6*10^3], L"\mathbf{2:1}", 10))
```

```julia
plt_modal = plot(legend=:topleft, xlabel="\nPolynomial degree \$p\$", xscale=:log10, yscale=:log10,
    ylabel="Spectral radius\n", fontfamily="Computer Modern", legendfontsize=10, ylims=[10^1, 2*10^7],
    xguidefontsize=12, yguidefontsize=12, xtickfontsize=12, ytickfontsize=12,
    xticks=([2,4,8,16,25], string.([2,4,8,16,25])),
    yticks=([10^2,10^4, 10^6], [L"10^2", L"10^4", L"10^6" ]), size=(450,300))

plot!(plt_modal, collect(2:16), s_mtu[1:15], label="Modal tensor-product",
    markershape=:square, markersize=4, linewidth=2.0, 
    markerstrokewidth=0.0, color=palette(:default)[2])
plot!(plt_modal, collect(2:16), s_mtc[1:15], label="", 
    linestyle=:dot, markershape=:square, markersize=4, linewidth=2, 
    markerstrokewidth=0.0, color=palette(:default)[2])
plot!(plt_modal, collect(2:10), s_mmu[1:9], label="Modal multidimensional", 
    markershape=:circle, markersize=3, linewidth=2, markerstrokewidth=0.0,
    markerstrokecolor=:slateblue4, color=:slateblue4)
plot!(plt_modal, collect(2:10), s_mmc[1:9], label="", 
    linestyle=:dot, markershape=:circle, markersize=3, 
    linewidth=2, markerstrokewidth=0.0, markerstrokecolor=:slateblue4, color=:slateblue4)

r1 = 12
r2 = 16
plot!(plt_modal, [r1, r2, r1, r1], 9.0*[r1^2, r2^2, r2^2, r1^2], 
    label=:none, linewidth=2, color=:black, 
    annotations = ([14.25], [4.0*10^3], L"\mathbf{2:1}", 10))
```

```julia
savefig(plt_nodal, "../plots/spectral_radius/spectral_radius_tet_nodal.pdf")
savefig(plt_modal, "../plots/spectral_radius/spectral_radius_tet_modal.pdf")
```
