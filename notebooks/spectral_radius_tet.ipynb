# Spectral radius for tetrahedra

```julia
using StableSpectralElements, JLD2, Plots, LaTeXStrings
```

```julia
path = "../results/spectral_radius/"

s_nmc = load_object(string(path,"NodalMulti_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_nmu = load_object(string(path,"NodalMulti_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_mmc = load_object(string(path,"ModalMulti_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_mmu = load_object(string(path,"ModalMulti_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_ntc = load_object(string(path,"NodalTensor_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_ntu = load_object(string(path,"NodalTensor_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_mtc = load_object(string(path,"ModalTensor_Tet_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_mtu = load_object(string(path,"ModalTensor_Tet_SkewSymmetricMapping/upwind/spectral_radii.jld2"));
```

```julia
plt_m_16 = plot(legend=:topleft, xlabel="Polynomial degree \$p\$", xscale=:log10, yscale=:log10,
    ylabel="Spectral radius", fontfamily="Computer Modern", legendfontsize=10, ylims=[5*10^1,10^3],
    xguidefontsize=12, yguidefontsize=12, xtickfontsize=12, ytickfontsize=12, 
    xticks=([2,4,8,16], string.([2,4,8,16])), yticks=([10^2,10^3], [L"10^2", L"10^3"]),
    size=(450,300))
plot!(plt_m_16, collect(2:10), s_nmu, label="Nodal multidimensional", 
    markershape=:circle, markersize=5, linewidth=2.5, markerstrokewidth=2.0, 
     markerstrokecolor=:slateblue4, color=:slateblue4)
plot!(plt_m_16, collect(2:10), s_nmc, label="", 
    linestyle=:dot, markershape=:circle, markersize=5, linewidth=2.5,
    markerstrokewidth=2.0,  markerstrokecolor=:slateblue4, color=:slateblue4)
plot!(plt_m_16, collect(2:10), s_mmu, label="Modal multidimensional",
    markershape=:circle, markersize=3, linewidth=2.0, 
    markerstrokewidth=0.0, color=palette(:default)[1])
plot!(plt_m_16, collect(2:10), s_mmc, label="", 
    linestyle=:dot, markershape=:circle, markersize=3, 
    linewidth=2.0, markerstrokewidth=0.0, color=palette(:default)[1])
r1 = 8
r2 = 10
plot!(plt_m_16, [r1, r2, r1, r1], 7*[r1^2, r2^2, r2^2, r1^2], 
        label=:none, linewidth=2, color=:black, annotations = ([9], [7.9*10^2], L"\mathbf{2:1}", 10))
```

```julia
plt_t_16 = plot(legend=:topleft, xlabel="Polynomial degree \$p\$", xscale=:log10, yscale=:log10,
    ylabel="Spectral radius", fontfamily="Computer Modern", legendfontsize=10, ylims=[10^1, 2*10^7],
    xguidefontsize=12, yguidefontsize=12, xtickfontsize=12, ytickfontsize=12,
    xticks=([2,4,8,16,25], string.([2,4,8,16,25])), size=(450,300), 
    yticks=([10^2,10^4, 10^6], [L"10^2", L"10^4", L"10^6" ]))
plot!(plt_t_16, collect(2:16), s_ntu, label="Nodal tensor-product", 
    markershape=:square, markersize=3, linewidth=2.5, 
    markerstrokewidth=0.0,  markerstrokecolor=:orangered4, color=:orangered4)
plot!(plt_t_16, collect(2:16), s_ntc, label="", 
    linestyle=:dot, markershape=:square, markersize=3, 
    linewidth=2.5, markerstrokewidth=0.0, markerstrokecolor=:orangered4, color=:orangered4)
plot!(plt_t_16, collect(2:16), s_mtu, label="Modal tensor-product", 
    markershape=:square, markersize=3, linewidth=2.0, markerstrokewidth=0.0, 
    markerstrokecolor=palette(:default)[2], color=palette(:default)[2])
plot!(plt_t_16, collect(2:16), s_mtc, label="",
    linestyle=:dot, markershape=:square, markersize=3, linewidth=2.0, 
    markerstrokewidth=0.0, markerstrokecolor=palette(:default)[2], 
    color=palette(:default)[2])
r1 = 12
r2 = 16
plot!(plt_t_16, [r1, r2, r1, r1], 9.0*[r1^2, r2^2, r2^2, r1^2], 
    label=:none, linewidth=2, color=:black, 
    annotations = ([14.25], [4.0*10^3], L"\mathbf{2:1}", 10))
```

```julia
savefig(plt_m_16, "../plots/spectral_radius/spectral_radius_tet_multi.pdf")
savefig(plt_t_16, "../plots/spectral_radius/spectral_radius_tet_tensor.pdf")
```
