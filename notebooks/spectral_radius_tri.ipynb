```julia
using CLOUD, JLD2, Plots, LaTeXStrings
```

```julia
path = "../results/spectral_radius/"

s_nmc = load_object(string(path,"NodalMulti_Tri_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_nmu = load_object(string(path,"NodalMulti_Tri_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_mmc = load_object(string(path,"ModalMulti_Tri_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_mmu = load_object(string(path,"ModalMulti_Tri_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_ntc = load_object(string(path,"NodalTensor_Tri_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_ntu = load_object(string(path,"NodalTensor_Tri_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
s_mtc = load_object(string(path,"ModalTensor_Tri_SkewSymmetricMapping/central/spectral_radii.jld2"))
s_mtu = load_object(string(path,"ModalTensor_Tri_SkewSymmetricMapping/upwind/spectral_radii.jld2"))
```

```julia
plt_m_16 = plot(legend=:topleft, xlabel="Polynomial degree \$p\$", xscale=:log10, yscale=:log10,
    ylabel="Spectral radius", fontfamily="Computer Modern", legendfontsize=9, ylims=[10^1,2*10^3],
    xguidefontsize=10, yguidefontsize=10, xtickfontsize=10, ytickfontsize=10,
xticks=([2,4,8,16,25], string.([2,4,8,16,25])), size=(400,300))
plot!(plt_m_16, collect(2:16), s_nmu[1:15], label="Nodal multidimensional",  markershape=:circle, markersize=5, linewidth=3.0, markerstrokewidth=0.0, color=:black)
plot!(plt_m_16, collect(2:16), s_nmc[1:15], label="", linestyle=:dash, markershape=:circle, markersize=5, linewidth=3.0, markerstrokewidth=0.0, color=:black)
plot!(plt_m_16, collect(2:16), s_mmu[1:15], label="Modal multidimensional",  markershape=:circle, markersize=3, linewidth=2.0, markerstrokewidth=0.0, color=palette(:default)[1])
plot!(plt_m_16, collect(2:16), s_mmc[1:15], label="", linestyle=:dash, markershape=:circle, markersize=3, linewidth=2.0, markerstrokewidth=0.0, color=palette(:default)[1])
r1 = 12
r2 = 16
plot!(plt_m_16, [r1, r2, r2, r1], 2.5*[r1^2, r2^2, r1^2, r1^2], label=:none, linewidth=2, color=:black, annotations = ([14.25], [2.9*10^2], L"\mathbf{2:1}", 10))
```

```julia
plt_t_16 = plot(legend=:topleft, xlabel="Polynomial degree \$p\$", xscale=:log10, yscale=:log10,
    ylabel="Spectral radius", fontfamily="Computer Modern", legendfontsize=9, ylims=[10^1, 10^6],
    xguidefontsize=10, yguidefontsize=10, xtickfontsize=10, ytickfontsize=10,
xticks=([2,4,8,16,25], string.([2,4,8,16,25])), size=(400,300))
plot!(plt_t_16, collect(2:16), s_ntu[1:15], label="Nodal tensor-product",  markershape=:square, markersize=3, linewidth=3.0, markerstrokewidth=0.0, color=:black)
plot!(plt_t_16, collect(2:16), s_ntc[1:15], label="", linestyle=:dash, markershape=:square, markersize=3, linewidth=3.0, markerstrokewidth=0.0, color=:black)
plot!(plt_t_16, collect(2:16), s_mtu[1:15], label="Modal tensor-product",  markershape=:square, markersize=3, linewidth=3.0, markerstrokewidth=0.0, markerstrokecolor=palette(:default)[2], color=palette(:default)[2])
plot!(plt_t_16, collect(2:16), s_mtc[1:15], label="", linestyle=:dash, markershape=:square, markersize=3, linewidth=3.0, markerstrokewidth=0.0, markerstrokecolor=palette(:default)[2], color=palette(:default)[2])
r1 = 12
r2 = 16
plot!(plt_t_16, [r1, r2, r2, r1], 2.2*[r1^2, r2^2, r1^2, r1^2], label=:none, linewidth=2, color=:black, annotations = ([14.25], [2*10^2], L"\mathbf{2:1}", 10))
plot!(plt_t_16, [r1, r2, r1, r1], 2.7*[r1^4, r2^4, r2^4, r1^4], label=:none, linewidth=2, color=:black, annotations = ([14.25], [2.8*10^5], L"\mathbf{4:1}", 10))
```

```julia
savefig(plt_m_16, "../plots/spectral_radius/spectral_radius_tri_multi_2_16.pdf")
savefig(plt_t_16, "../plots/spectral_radius/spectral_radius_tri_tensor_2_16.pdf")
```
